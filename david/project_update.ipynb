```python
from pynq import Overlay
import numpy as np
import random
from pynq import allocate

stream_overlay = Overlay("/home/xilinx/pynq/overlays/maxVal/maxValue.bit")
SAMPLES = 601
```

```python
dma = stream_overlay.maxDMA.axi_dma_0
input_buffer = allocate(shape=(SAMPLES,), dtype=np.int16)
output_buffer = allocate(shape=(SAMPLES,), dtype=np.int16)
```

```python
input_buffer[:] = [random.randint(0,100) for k in range(SAMPLES)]
print(input_buffer)
```

```python
def run_kernel():
    dma.sendchannel.transfer(input_buffer)
    dma.recvchannel.transfer(output_buffer)
    dma.sendchannel.wait()
    dma.recvchannel.wait()
```

```python
run_kernel()
```
